```python
import numpy as np

# ============================================================
# Funções auxiliares
# ============================================================

def iou(mask1, mask2):
    inter = np.logical_and(mask1, mask2).sum()
    union = np.logical_or(mask1, mask2).sum()
    return inter / union if union > 0 else 0.0

def box_iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    inter = max(0, xB - xA) * max(0, yB - yA)
    areaA = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    areaB = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])

    union = areaA + areaB - inter
    return inter / union if union > 0 else 0.0


# ============================================================
# MÉTRICAS
# ============================================================

def compute_metrics(pred_masks, gt_masks, pred_boxes, gt_boxes):
    metrics = {}

    # 1) mIoU de máscaras
    ious = []
    for pm, gm in zip(pred_masks, gt_masks):
        ious.append(iou(pm > 0, gm > 0))
    metrics["mIoU_masks"] = float(np.mean(ious))

    # 2) IoU médio das caixas
    box_ious = []
    for pb, gb in zip(pred_boxes, gt_boxes):
        box_ious.append(box_iou(pb, gb))
    metrics["mIoU_boxes"] = float(np.mean(box_ious))

    # 3) Precisão / Recall com IoU > 0.5
    TP = sum(i > 0.5 for i in box_ious)
    FP = max(0, len(pred_boxes) - TP)
    FN = max(0, len(gt_boxes) - TP)

    metrics["Precision"] = TP / (TP + FP + 1e-6)
    metrics["Recall"] = TP / (TP + FN + 1e-6)

    # 4) mAP simples (apenas para IoU>0.5)
    metrics["mAP50"] = metrics["Precision"]

    return metrics


# ============================================================
# EXEMPLO DE USO
# ============================================================

# pred_masks  => liste suas máscaras aqui (binárias)
# pred_boxes  => caixas previstas
# gt_masks    => máscaras GT
# gt_boxes    => caixas GT

# ⚠️ Exemplo fictício (substitua pelos seus)
pred_masks = [np.random.rand(256,256) > 0.8]
gt_masks   = [np.random.rand(256,256) > 0.8]

pred_boxes = [np.array([50,50,200,200])]
gt_boxes   = [np.array([60,60,210,210])]

# Calcula
metrics = compute_metrics(pred_masks, gt_masks, pred_boxes, gt_boxes)

# Imprime tudo bonito
print("\n===== MÉTRICAS DO MODELO =====")
for k,v in metrics.items():
    print(f"{k:15s}: {v:.4f}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image

# ========== 1. IMAGEM ORIGINAL ==========
orig = Image.open(img_path).convert("RGB")
orig_np = np.array(orig)
H, W = orig_np.shape[:2]
vars(outputs)
attrs = vars(outputs)

# tenta achar máscaras
mask_key = None
for k in attrs.keys():
    if "mask" in k.lower():
        mask_key = k
        break

# tenta achar caixas
box_key = None
for k in attrs.keys():
    if "box" in k.lower():
        box_key = k
        break

print("Atributos encontrados:", attrs.keys())
print(">> máscara =", mask_key, "| caixas =", box_key)

pred_masks = attrs.get(mask_key, None)
pred_boxes = attrs.get(box_key, None)

# ========== 3. PROCESSAR MÁSCARAS ==========
if pred_masks is None:
    raise ValueError("Nenhuma máscara encontrada nos outputs.")

# pred_masks: [1, num_queries, H/4, W/4]
masks = pred_masks.sigmoid() > 0.5
masks = masks[0].detach().cpu().numpy()

# upscale para tamanho original
masks_up = []
for m in masks:
    m_img = Image.fromarray((m * 255).astype("uint8"))
    m_img = m_img.resize((W, H), Image.NEAREST)
    masks_up.append(np.array(m_img) > 128)

masks_up = np.stack(masks_up)

# ========== 4. CRIAR OVERLAY DE COR ==========
combined_mask = np.zeros((H, W, 3), dtype=np.uint8)

for m in masks_up:
    color = np.random.randint(0, 255, size=3)
    combined_mask[m] = color

overlay = (0.6 * orig_np + 0.4 * combined_mask).astype(np.uint8)

# ========== 5. DESENHAR CAIXAS SE EXISTIREM ==========
if pred_boxes is not None:
    boxes = pred_boxes[0].detach().cpu().numpy()
    for b in boxes:
        x1, y1, x2, y2 = b.astype(int)
        cv2.rectangle(overlay, (x1, y1), (x2, y2), (255, 0, 0), 2)

# ========== 6. MOSTRAR RESULTADO ==========
plt.figure(figsize=(10, 10))
plt.imshow(overlay)
plt.axis("off")
plt.show()
```

# Carregar as imagens de teste

```python
import os
from PIL import Image
from torchvision import transforms

def load_image(image_path):
    image = Image.open(image_path).convert("RGB")
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    return transform(image).unsqueeze(0)  # Adiciona uma dimensão extra para o batch size
```

```python
from pycocotools.coco import COCO

ImgNm = "Art1P1"
# Caminho para o arquivo de anotações COCO
annotation_file = f"data/Layout/MPU/data/MPU/train/coco/{ImgNm}.json"

coco = COCO(annotation_file)

# Exemplo de como pegar as IDs das imagens para fazer as predições
img_ids = coco.getImgIds()

# Exemplo de carregar uma imagem e suas anotações
image_info = coco.loadImgs(img_ids[0])[0]
image = load_image(f"data/Layout/MPU/data/MPU/train/image/{ImgNm}.png")
```

```python
model.load_state_dict(torch.load("pretrained_model/docsam_large_doclaynet.pth", map_location="cpu"))
model.eval()   # 🔥 IMPORTANTE
```

```python
with torch.no_grad():  # Não precisamos de gradientes para a inferência
    prediction = model(image)  # Faz a predição

# A predição contém vários resultados (como caixas delimitadoras, máscaras, etc.)
masks = prediction[0]['masks']  # Máscaras segmentadas
boxes = prediction[0]['boxes']  # Caixas delimitadoras
labels = prediction[0]['labels']  # Rótulos das classes
scores = prediction[0]['scores']  # Scores de confiança
```
